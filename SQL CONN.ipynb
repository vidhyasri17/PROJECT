```python

#importing libraries
import pandas as pd
import psycopg2
import numpy as np
```

```python
# csv to dataframe
df_buses_1=pd.read_csv("df_buses_1.csv")
df_buses_2=pd.read_csv("df_buses_2.csv")
df_buses_3=pd.read_csv("df_buses_3.csv")
df_buses_4=pd.read_csv("df_buses_4.csv")
df_buses_5=pd.read_csv("df_buses_R.csv")
df_buses_6=pd.read_csv("df_buses_SB.csv")
df_buses_7=pd.read_csv("df_buses_H.csv")
df_buses_8=pd.read_csv("df_buses_AS.csv")
df_buses_9=pd.read_csv("df_buses_9.csv")
df_buses_10=pd.read_csv("df_buses_10.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

# DATA CLEANING

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0) 
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True) 
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
import psycopg2

def create_bus_details_table():
    try:
        # Establish connection
        conn = psycopg2.connect(
            host="localhost",
            user="postgres",
            port="5432",
            database="redbus_data",
            password="05.svidhya.17"
        )

        # Enable autocommit
        conn.autocommit = True

        cursor = conn.cursor()

        drop_query='''DROP TABLE IF EXISTS bus_details;'''
        cursor.execute(drop_query)
        conn.commit()

        # SQL query to create the table
        create_query = '''
        CREATE TABLE IF NOT EXISTS bus_details (
                        ID SERIAL PRIMARY KEY,
                        Bus_name TEXT NOT NULL,
                        Bus_type TEXT NOT NULL,
                        Start_time TIMESTAMP NOT NULL,
                        End_time TIMESTAMP NOT NULL,
                        Total_duration TEXT NOT NULL,
                        Price DECIMAL NOT NULL,
                        Seats_Available TEXT NOT NULL,
                        Ratings FLOAT NULL,
                        Route_link TEXT NULL,
                        Route_name TEXT NULL
        )
        '''

        # Execute the create table query
        cursor.execute(create_query)
        print("Table created successfully")

    except psycopg2.Error as e:
        print(f"An error occurred: {e}")

    finally:
        # Close the cursor and connection
        if 'cursor' in locals():
            cursor.close()
        if 'conn' in locals():
            conn.close()

# Execute the function to create the table
create_bus_details_table()
```

```python
import pandas as pd
import psycopg2



# Ensure datetime format for Start_time and End_time
Final_df['Start_time'] = pd.to_datetime(Final_df['Start_time'])
Final_df['End_time'] = pd.to_datetime(Final_df['End_time'])

# Define the function
def insert_data_into_bus_details(Final_df):
    Final_df['Bus_type'] = Final_df['Bus_type'].fillna('Unknown')
    Final_df['Seats_Available'] = Final_df['Seats_Available'].fillna('Unknown')

    conn = None
    my_cursor = None

    try:
        conn = psycopg2.connect(
            host="localhost",
            user="postgres",
            port="5432",
            database="redbus_data",
            password="05.svidhya.17"  # Replace with your actual password
        )
        my_cursor = conn.cursor()

        insert_query = '''
        INSERT INTO bus_details(
            Bus_name,
            Bus_type,
            Start_time,
            End_time,
            Total_duration,
            Price,
            Seats_Available,
            Ratings,
            Route_link,
            Route_name)
        VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        '''

        data = Final_df[['Bus_name', 'Bus_type', 'Start_time', 'End_time', 'Total_duration', 'Price', 'Seats_Available', 'Ratings', 'Route_link', 'Route_name']].values.tolist()

        for row in data:
            my_cursor.execute(insert_query, row)

        conn.commit()
        print("Values inserted successfully")

    except psycopg2.Error as e:
        print(f"An error occurred: {e}")

    finally:
        if my_cursor:
            my_cursor.close()
        if conn:
            conn.close()

# Call the function
insert_data_into_bus_details(Final_df)
```
